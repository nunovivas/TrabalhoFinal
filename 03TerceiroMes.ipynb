## Visualização

### Matplotlib

O **Matplotlib** é uma biblioteca poderosa de visualização de dados para Python, amplamente utilizada para criar gráficos 2D e até mesmo alguns 3D. É uma ferramenta indispensável na análise de dados, pois permite transformar grandes conjuntos de informações em gráficos claros e interpretáveis.

Com o **Matplotlib**, pode criar diversos tipos de gráficos, como linhas, barras, dispersão (scatter), histogramas, entre outros. A biblioteca é altamente personalizável, permitindo ajustar cores, legendas, títulos, etiquetas dos eixos e muito mais. Além disso, é frequentemente usada em conjunto com outras bibliotecas, como **NumPy** e **Pandas**, para visualização de dados mais complexos.

Um dos módulos mais conhecidos do Matplotlib é o **pyplot**, que oferece uma interface simples e intuitiva, semelhante à biblioteca gráfica do MATLAB. Isso torna o Matplotlib acessível mesmo para iniciantes.

A biblioteca é ideal para criar visualizações de alta qualidade para relatórios ou apresentações. Contudo, como é bastante detalhada, pode ser um pouco mais trabalhosa para configurar em comparação com outras bibliotecas como o Seaborn. 

Abaixo encontra-se um exemplo básico para criar um gráfico de linhas:


````python
import matplotlib.pyplot as plt

# Dados
anos = [2020, 2021, 2022, 2023]
vendas = [250, 300, 350, 400]

# Criar o gráfico
plt.plot(anos, vendas, marker='o', color='blue', label='Vendas')

# Personalização
plt.title("Vendas Anuais")
plt.xlabel("Ano")
plt.ylabel("Vendas (em milhares)")
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
````
